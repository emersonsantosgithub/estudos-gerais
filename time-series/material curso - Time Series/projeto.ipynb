## ANALISANDO VENDA POR MES DA EMPRESA ALUCAR 

**ANÁLISE FOCADA EM TIME SERIES**

```python
#BIBLIOTECA
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

#BASE 
alucar = pd.read_csv("alucar.csv")
```

```python
# TRANSFORMAÇÃO DE OBJETC PARA DATA
alucar['mes'] = pd.to_datetime(alucar['mes'])
```

**QUANTIDADE DE VENDA ENTRE ANO MES**

```python
plt.figure(figsize=(10, 4))

sns.lineplot(data=alucar, x='mes', y='vendas')
plt.title("Alucar:Vendas por mês entre 2017 á 2019", loc='left', fontsize=12)
plt.show()
```

**DIFERENCA ENTRE MES ATUAL E MES ANTERIOR DAS VENDAS**

```python
alucar['Aumento'] = alucar['vendas'].diff()
```

```python
# FUNCAO PARA PLOTAGEM
def plotagem(dataset, eixo_x, eixo_y, titulo):
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=dataset, x=eixo_x, y=eixo_y)
    plt.title(titulo, loc='left', fontsize=12)
    plt.show()

    
```

```python
plotagem(dataset=alucar, eixo_x='mes', eixo_y='Aumento', titulo='Alucar:Venda por mês entre 2017 á 2019')
```

**DIFERENCA EM VENDA  AUMENTO DO MES ATUAL E ANTERIOR**

```python
alucar['Aceleracao'] = alucar['Aumento'].diff()
alucar.head(5)
```

```python
plotagem(dataset=alucar, eixo_x='mes', eixo_y='Aceleracao', titulo="Aceleracao")
```

```python


plt.figure(figsize=(16, 12))
sns.set_style("darkgrid")  # Isso irá aplicar um fundo escuro e grid claro por padrão
# Bloco 1
ax1 = plt.subplot(3, 1, 1)
sns.lineplot(data=alucar, x='mes', y='vendas', ax=ax1)
ax1.set_title('Vendas por mês')

# Bloco 2
ax2 = plt.subplot(3, 1, 2)
sns.lineplot(data=alucar, x='mes', y='Aumento', ax=ax2)
ax2.set_title('Aumento por mês')

# Bloco 3
ax3 = plt.subplot(3, 1, 3)
sns.lineplot(data=alucar, x='mes', y='Aceleracao', ax=ax3)
ax3.set_title('Aceleração por mês')

plt.tight_layout()  # Ajusta automaticamente os espaços entre os subgráficos
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
plt.figure(figsize=(12,4))

autocorrelation_plot(alucar['vendas'])
plt.title("Autocorrelacao com vendas", loc='left', fontsize=12)
plt.show()

```

```python
plt.figure(figsize=(12,4))

autocorrelation_plot(alucar['Aumento'][1:])
plt.title("Autocorrelacao com Aumento", loc='left', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,4))

autocorrelation_plot(alucar['Aceleracao'][2:])
plt.title("Autocorrelacao com Aceleracao", loc='left', fontsize=12)
plt.show()
```
