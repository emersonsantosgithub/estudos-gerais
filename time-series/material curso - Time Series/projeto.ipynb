```python
import pandas as pd 
```

```python
alucar = pd.read_csv("alucar.csv")
```

```python
alucar['mes'] = pd.to_datetime(alucar['mes'])
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 4))

sns.lineplot(data=alucar, x='mes', y='vendas')
plt.title("Alucar:Vendas por mês entre 2017 á 2019", loc='left', fontsize=12)
plt.show()
```

```python
alucar['Aumento'] = alucar['vendas'].diff()
```

```python
def plotagem(dataset, eixo_x, eixo_y, titulo):
    plt.figure(figsize=(10, 4))
    sns.lineplot(data=dataset, x=eixo_x, y=eixo_y)
    plt.title(titulo, loc='left', fontsize=12)
    plt.show()

    
```

```python
plotagem(dataset=alucar, eixo_x='mes', eixo_y='Aumento', titulo='Alucar:Venda por mês entre 2017 á 2019')
```

```python
alucar['Aceleracao'] = alucar['Aumento'].diff()
alucar.head(5)
```

```python
plotagem(dataset=alucar, eixo_x='mes', eixo_y='Aceleracao', titulo="Acelelaracao")
```
